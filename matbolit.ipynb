```python
import matplotlib.pyplot as plt
import pandas as pd
#scatter : nuage de points
#plot : diagramme en ligne
#bar : diagramme en barres
#his : histogramme
#pie : diagramme circulaire
prets = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/prets.csv')
prets['taux_endettement'] = round((prets['remboursement']*100)/prets['revenu'],2) 
plt.scatter(prets['revenu'] , prets['taux_endettement'] , s = 60 , color = "red" , ) #nuage de points
```

```python
data = prets.groupby('ville')['remboursement'].sum()
data = data.reset_index()
```

```python
plt.pie(x = data['remboursement'] , labels = data['ville'])
#pour afficher le pourcentage
plt.pie(x = data['remboursement'] , labels = data['ville'] , autopct ="%.2f%%")
```

```python
#plt.bar(height=data['remboursement'] ,  x = data['ville'])
data_sorted = data.sort_values('remboursement' , ascending = False)
plt.bar(height=data_sorted['remboursement'] ,  x = data_sorted['ville'])
```

```python
evolution = pd.DataFrame({'date' : ['2013-02-01' ,  '2014-05-03' , '2015-03-01' ,'2016-01-04'] , 'chiffre' : [183000 , 199000 ,175000 , 217000]
                         })
plt.plot(evolution['date'] , evolution['chiffre'])
```

```python
df1 = prets.loc[prets['ville'] == 'TOULOUSE']
df2 = prets.loc[prets['ville'] == 'PARIS']
plt.scatter(df1['revenu'] , df1['taux'] , label = "TOULOUSE")
plt.scatter(df2['revenu'] , df2['taux'] , label = "PARIS")
plt.legend()
```

```python
#boucle pour afficher tous les viles
for i in prets['ville'].unique():
    df = prets.loc[prets['ville'] == i , :]
    plt.scatter(df['remboursement'] , df['taux'] , label = i)
plt.legend()
plt.show()
```
