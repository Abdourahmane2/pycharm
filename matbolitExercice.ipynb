```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#exercice
prets = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/prets_final.csv')
#propotion de pret par type de prets
data = prets.groupby('type').size() #compter le nombre 
data = data.reset_index()
data.columns = ['type' , 'nombre'] #renomer les colones
plt.bar(height = data['nombre'] , x = data['type']) 
```

```python
#benefices realise en fonction du revenu client type immobilier
bene = prets.loc[prets['type'] =="immobilier" , :]
plt.scatter(bene['revenu'] , bene['benefices'])
```

```python
#distribution des benefices realise
plt.hist(prets['benefices'])
```

```python
bene_ville = prets.groupby('ville')['benefices'].sum()
bene_ville = bene_ville.reset_index()
plt.bar(height = bene_ville['benefices'] , x = bene_ville["ville"] , color = "red" ) 
```
